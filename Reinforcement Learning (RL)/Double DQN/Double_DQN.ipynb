```python
import numpy as np

Q_online = np.random.rand(5,2)
Q_target = np.random.rand(5,2)

state = 2
next_state = 3
reward = 1
gamma = 0.9

# Action selection from online network
best_action = np.argmax(Q_online[next_state])

# Evaluation from target netwrok
target_value = reward + gamma * Q_target[next_state][best_action] 

Q_online[state][best_action] = target_value 

print(Q_online) 
```
