```python
import pandas as pd 
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.naive_bayes import MultinomialNB

# Step 1: Create Dataset 
data = {
    'Message':[
        'Win money now', 
        'Limited offer free prize',
        'Meeting schedule tomorrow',
        'Project discussion today'
    ],
    'Spam':[1,1,0,0]
}

df = pd.DataFrame(data)

# Step 2: Convert text into numerical features 
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['Message'])

y = df['Spam']

# Step 3: Train Naive Bayes Model
model = MultinomialNB()
model.fit(X,y)

# Step 4: Predict new email 
new_email = vectorizer.transform(['Free money offer'])
prediction = model.predict(new_email)
```
