```python
import pandas as pd 
impprt matplotlib.pyplot as plt 
from sklearn.cluster import KMeans

# Step 1: Create Dataset 
data = {
    'Age':[22,25,30,35,40,28],
    'Income':[25,30,50,60,80,45]
}

df = pd.DataFrame(data)

X = df[['Age', 'Income']]

# step 2: Train K-Means
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)

# Step 3: Cluster Labels
df['Cluster'] = kmeans.labels_

print(df)

# Step 4: Visualization 
plt.scatter(df['Age'], df['Income'], c=df['Cluster'])
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Customer Segmentation using K
```
